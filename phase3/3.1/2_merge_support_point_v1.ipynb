```python
road_df = pd.read_csv(road_data_path+"/route_network_with_cluster.csv")

cluster_df = road_df[['cluster', 'center_lat', 'center_lon']]
cluster_df = cluster_df.drop_duplicates()

cluster_for_compare = cluster_df[['center_lat', 'center_lon']].apply(lambda x: round(x, 3)).drop_duplicates(['center_lat','center_lon'])
cluster_for_compare['cluster'] = cluster_df['cluster']
##########

days = range(1,32)
month = 3

# for month in months:

for day in days:

  filename = trips_data_path + f'/trips_2019_{month}_{day}.csv'

  try:
    trips_df = pd.read_csv(filename, parse_dates=['time_stamp'])
  except:
    pass

  trips_for_compare = trips_df[['lat', 'lon']].apply(lambda x: round(x, 3))

  trips_mask = trips_for_compare.merge(cluster_for_compare, left_on=['lat','lon'], right_on=['center_lat', 'center_lon'], how='left')[['lat', 'lon', 'cluster']]

  trips_mask = trips_mask.fillna(-1)
  trips_mask['cluster'] = trips_mask['cluster'].astype(int)

  trips_df['cluster'] = trips_mask['cluster']


  trips_df.to_csv(export_data_path + f'/trips_2019_{month}_{day}.csv', encoding='utf-8', index = False)
```
