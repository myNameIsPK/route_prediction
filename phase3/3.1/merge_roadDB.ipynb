#find rd of roadDB

```python
rd_of_roadDB = []
for cood in sp['lat-lon']:
  if roadDB_NE.loc[roadDB_NE['lat-lon'].apply(lambda x: True if x == cood else False)].empty:
    rd_of_roadDB.append(False)
  else:
    rd_of_roadDB.append(roadDB_NE.loc[roadDB_NE['lat-lon'].apply(lambda x: True if x == cood else False)].reset_index().rd[0])
```
