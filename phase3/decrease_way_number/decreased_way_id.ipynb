# decrease way_id data

```python
dfs = []
day_len = []
day_trip_count = []
days = range(1,32)
# days = range(1,2)
# months = range(1,2)
months = range(1,13)

for month in months:
  for day in days:
    path = trips_data_path + f'/way_labeled_2019_{month}_{day}.csv'
    try:
      trips_df = pd.read_csv(path, parse_dates=['time_stamp'])
    except:
      print("Fail!!", end="\t")
      print(path)
      continue
    print(f"{month}_{day}", end="\t")
    print(len(trips_df), end="\t")

    export_df = trips_df[(trips_df["label"]!=-1) & (trips_df["dist"] < 0.005)]
    export_df = export_df.drop_duplicates(subset=["trip_id", "way_id"])
    print(len(export_df))
    export_df.to_csv(export_data_path+f'/way_labeled_decrease_2019_{month}_{day}.csv', index=False)
```
