# Map trip to way sequence with ckd-tree

## load way segment

```python
label_map_df = pd.read_csv(road_data_path + '/wayid_label_mapped.csv')
label_map_df = label_map_df.drop_duplicates('way_id')
label_map_df
```

## load trips data

```python
dfs = []
day_len = []
day_trip_count = []
days = range(1,32)
# days = range(1,2)
# months = range(1,2)
months = range(1,12)

for month in months:
  for day in days:
    path = trips_data_path + f'/way_seq_2019_{month}_{day}.csv'
    try:
      trips_df = pd.read_csv(path, parse_dates=['time_stamp'])
    except:
      print("Fail!!", end="\t")
      print(path)
      continue
    print(f"{month}_{day}", end="\t")
    print(len(trips_df))

    export_df = trips_df.merge(label_map_df[['way_id', 'label']], left_on='way_id', right_on='way_id', how='left')
    export_df
    export_df.to_csv(export_data_path+f'/way_labeled_2019_{month}_{day}.csv', index=False)
```
