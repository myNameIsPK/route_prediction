## function

```python
from geopy.distance import great_circle as distance
from geopy.point import Point as Point
from math import sin, cos, atan2, sqrt, degrees, radians, pi, inf

def distance_km(location1, location2):
  lat1, lon1, ts1 = location1
  lat2, lon2, ts2 = location2
  radius = 6371 # km

  dlat = radians(lat2-lat1)
  dlon = radians(lon2-lon1)
  a = sin(dlat/2) * sin(dlat/2) + cos(radians(lat1)) \
      * cos(radians(lat2)) * sin(dlon/2) * sin(dlon/2)
  c = 2 * atan2(sqrt(a), sqrt(1-a))
  dist = radius * c

  # print('dist:'+str(dist))
  return round(dist)

def duration_min(location1, location2):
  lat1, lon1, ts1 = location1
  lat2, lon2, ts2 = location2
  diff = ts1 - ts2
  return round(diff.seconds / 60)
```

## operate

```python
import itertools

# for day in range(12,24):
month = '4'
for day in ['09']:
  print(f'day: {day}')
  trips_df = pd.read_csv(data_path + f'/trips_2019_{month}_{day}.csv', parse_dates=['time_stamp'])

  MAX_DIST = 10 # km
  MAX_DURA = 30 # min

  trips = trips_df.trip_id.value_counts(ascending=True)
  lenght = len(trips.index)
  count = 0
  false_trip = []
  jump_trip = []
  stay_trip = []

  for t_id in trips.index:
    count = count+1
    # print(f'id:{t_id}\t{count}\t/{lenght}')
    trip = trips_df[trips_df.trip_id == t_id]
    points = [(lat, lon, ts) for lat, lon, ts in list(zip(trip.lat, trip.lon, trip.time_stamp))]

    # stop = stops_df[stops_df.stop_id == trip.iloc[0].stop_id].iloc[0]
    # stop_p = (stop.lat, stop.lon, stop.time_stamp)
    # points.append(stop_p)
    
    comparable = zip(points[1:], points[:-1])
    for p1, p2 in zip(points[1:], points[:-1]):
      dist = distance_km(p1, p2)
      dura = duration_min(p1, p2)
      if dist > MAX_DIST and dura > MAX_DURA:
        # print('VANISH')
        # print(f'id:{t_id}\t{count}\t/{lenght}')
        false_trip.append(t_id)
        # print('dist:\t',end='')
        # print(*list(itertools.starmap(distance_km, zip(points[1:], points[:-1]))), sep='\t')
        # print('dura:\t',end='')
        # print(*list(itertools.starmap(duration_min, zip(points[1:], points[:-1]))), sep='\t')
        break
      if dura > MAX_DURA and dist <= MAX_DIST:
        # print('STAY')
        # print(f'id:{t_id}\t{count}\t/{lenght}')
        stay_trip.append(t_id)
        # print('dist:\t',end='')
        # print(*list(itertools.starmap(distance_km, zip(points[1:], points[:-1]))), sep='\t')
        # print('dura:\t',end='')
        # print(*list(itertools.starmap(duration_min, zip(points[1:], points[:-1]))), sep='\t')
        break
      if dist > MAX_DIST and dura <= MAX_DURA:
        # print('JUMP')
        # print(f'id:{t_id}\t{count}\t/{lenght}')
        jump_trip.append(t_id)
        # print('dist:\t',end='')
        # print(*list(itertools.starmap(distance_km, zip(points[1:], points[:-1]))), sep='\t')
        # print('dura:\t',end='')
        # print(*list(itertools.starmap(duration_min, zip(points[1:], points[:-1]))), sep='\t')
        break
  bad_msk = false_trip+jump_trip+stay_trip

  good_trips = trips_df[~trips_df.trip_id.isin(bad_msk)]
  bad_trips = trips_df[trips_df.trip_id.isin(bad_msk)]

  # good_trips.to_csv(data_path + f'/bangkok_trips_and_stops/{day}/clean_trip.csv', encoding='utf-8', index=False)
```
