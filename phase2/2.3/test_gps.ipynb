```python
import glob
import os
import re
# days = range(8,13)
days = [8]
# months = range(1,13)
months = [4]
filenames = glob.glob(data_path+f'/2019-*.csv')
filenames = [f for f in filenames if re.search(re.compile('/2019-0?('+'|'.join(str(i) for i in months)+')-0?(' + '|'.join(str(i) for i in days) + ')_[0-2][0-9]\.csv$'), f)]
filenames = sorted(filenames)
for f in sorted(filenames):
    print(f)
```

```python
dfs = [pd.read_csv(filename, parse_dates=['time_stamp']).sort_values(by='time_stamp',ignore_index=True) for filename in filenames]
df = pd.concat(dfs, ignore_index=True)
df
```

มีรถกี่คัน

```python
df['unit_id'].nunique()
```

```python
veh_count = df['unit_id'].value_counts()
veh_count[veh_count > 1000]
```

mean of vehicles record

```python
veh_count.mean()
```

```python
veh_count[veh_count < 100]
```

```python
fig, axs = plt.subplots(1, 5, sharey=True, tight_layout=True)
axs[0].hist(veh_count[veh_count < 100])
axs[1].hist(veh_count[(veh_count >= 100) & (veh_count < 200)])
axs[2].hist(veh_count[(veh_count >= 200) & (veh_count < 300)])
axs[3].hist(veh_count[(veh_count >= 500) & (veh_count < 1000)])
axs[4].hist(veh_count[veh_count >= 1000])
```
